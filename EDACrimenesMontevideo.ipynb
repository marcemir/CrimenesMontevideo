### CRÍMENES MONTEVIDEO

#### 0. DICIONARIO DE DATOS
#### Variables Geográficas y Temporales
- **`BARRIO_MONTEVIDEO`**: Nombre del barrio de Montevideo donde se registró el dato. _Tipo de dato_: Categórica. _Ejemplo_: `Pocitos`, `Ciudad Vieja`.
- **`POBLACION`**: Cantidad de población en el barrio correspondiente. _Tipo de dato_: Numérico (entero). _Ejemplo_: `5000`, `25000`.
- **`TURISTICO`**: Indica si el barrio es turístico. _Tipo de dato_: Binario (`1`=Sí, `0`=No). _Ejemplo_: `1`, `0`.
- **`FECHA`**: Fecha completa del registro. _Tipo de dato_: Fecha. _Formato_: `AAAA-MM-DD`. _Ejemplo_: `2023-01-01`.
- **`AÑO`**: Año del registro. _Tipo de dato_: Numérico (entero). _Ejemplo_: `2023`.
- **`MES`**: Número del mes en el que ocurrió el evento. _Tipo de dato_: Numérico (entero). _Ejemplo_: `1` (Enero), `12` (Diciembre).
- **`DIA_MES`**: Día del mes en el que ocurrió el evento. _Tipo de dato_: Numérico (entero). _Ejemplo_: `1`, `15`, `31`.
- **`DIA_SEMANA`**: Nombre del día de la semana. _Tipo de dato_: Categórica. _Ejemplo_: `Lunes`, `Martes`.
- **`SEMANA_MES`**: Número de la semana dentro del mes. _Tipo de dato_: Numérico (entero). _Ejemplo_: `1`, `2`, `3`.
- **`SEMESTRE`**: Número del semestre del año. _Tipo de dato_: Numérico (entero). _Ejemplo_: `1` (Primer Semestre), `2` (Segundo Semestre).
- **`TRIMESTRE`**: Número del trimestre del año. _Tipo de dato_: Numérico (entero). _Ejemplo_: `1`, `2`, `3`.

#### Variables Climáticas
- **`TEMPMAX`**: Temperatura máxima registrada en el día (en °C). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `25.5`, `30.1`.
- **`TEMPMIN`**: Temperatura mínima registrada en el día (en °C). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `15.2`, `18.0`.
- **`TEMP`**: Temperatura promedio registrada en el día (en °C). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `20.3`, `22.8`.
- **`FEELSLIKEMAX`**: Sensación térmica máxima registrada en el día (en °C). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `28.5`, `35.0`.
- **`FEELSLIKEMIN`**: Sensación térmica mínima registrada en el día (en °C). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `12.5`, `15.0`.
- **`FEELSLIKE`**: Sensación térmica promedio registrada en el día (en °C). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `20.0`, `24.0`.
- **`HUMIDITY`**: Humedad promedio en el día (en %). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `60`, `85`.
- **`PRECIP`**: Volumen de lluvia registrado en el día (en mm). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `0.0`, `12.5`, `50.0`.
- **`WINDGUST`**: Velocidad máxima de las rachas de viento (en km/h). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `30.0`, `50.0`.
- **`WINDSPEED`**: Velocidad promedio del viento durante el día (en km/h). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `10.5`, `20.0`.
- **`SEALEVELPRESSURE`**: Presión atmosférica promedio registrada en el día (en mbar). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `1013`, `1018`.
- **`VISIBILITY`**: Visibilidad promedio en el día (en km). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `10.0`, `5.0`.
- **`SEVERERISK`**: Nivel de alerta por riesgo severo en una escala de 0 a 100. _Tipo de dato_: Numérico (entero). _Ejemplo_: `15` (bajo), `85` (alto).
- **`SEVERERISKCAT`**: Categoría del riesgo severo según el nivel (`Bajo`, `Medio`, `Alto`). _Tipo de dato_: Categórica. _Ejemplo_: `Bajo`, `Medio`, `Alto`.
- **`CONDITIONCLEAR`**: Indica si el día fue despejado. _Tipo de dato_: Binario (`1`=Sí, `0`=No). _Ejemplo_: `1`, `0`.
- **`CONDITIONPARTIALLYCLOUDY`**: Indica si el día fue parcialmente nublado. _Tipo de dato_: Binario (`1`=Sí, `0`=No). _Ejemplo_: `1`, `0`.
- **`CONDITIONOVERCAST`**: Indica si el día fue totalmente cubierto. _Tipo de dato_: Binario (`1`=Sí, `0`=No). _Ejemplo_: `1`, `0`.
- **`CONDITIONRAIN`**: Indica si el día fue lluvioso. _Tipo de dato_: Binario (`1`=Sí, `0`=No). _Ejemplo_: `1`, `0`.
- **`WINDIRCAT`**: Dirección del viento categorizada. _Tipo de dato_: Categórica. _Ejemplo_: `N`, `S`, `E`, `O`.

#### Variables Económicas y Sociales
- **`TASA_DESEMPLEO`**: Nivel de desempleo en Montevideo durante el período (en %). _Tipo de dato_: Numérico (flotante). _Ejemplo_: `7.5`, `10.2`.

#### Variables de Festivos
- **`NO_LABORABLE`**: Indica si el día es feriado no laborable. _Tipo de dato_: Categórica (`1`=Sí, `0`=No, `999`=No corresponde). _Ejemplo_: `1`, `0`, `999`.
- **`NOMBRE_FESTIVO`**: Nombre del día festivo si corresponde. _Tipo de dato_: Categórica. _Ejemplo_: `Año Nuevo`, `Navidad`.

#### Variables de Crímenes
- **`COUNT_DELITOS`**: Cantidad de delitos denunciados en la agrupación correspondiente. _Tipo de dato_: Numérico (entero). _Ejemplo_: `0`, `15`, `50`.
- **`COUNT_DELITOS_NORM`**: Cantidad de delitos normalizada por la población del barrio. _Tipo de dato_: Numérico (flotante). _Ejemplo_: `0.5`, `2.8`, `15.0`.
- **`RIESGO`**: **Variable objetivo** que clasifica el riesgo de delito en una agrupación. _Tipo de dato_: Categórica. _Ejemplo_: `Bajo`, `Alto`.

#### 2.2. ANÁLISIS BIVARIADO

##### 2.2.1. VARIABLES CATEGÓRICAS

```python
# Grafico de Frecuencia registros por tipo de riesgo y barrio

# Ordenar los barrios según la cantidad de delitos de alto riesgo de menos a más
orden_barrio = df[df['RIESGO'] == 'Alto']['BARRIO_MONTEVIDEO'].value_counts(ascending=True).index

# Crear el gráfico con el orden personalizado
plt.figure(figsize=(10, 20))
sns.countplot(
    y='BARRIO_MONTEVIDEO', 
    hue='RIESGO', 
    data=df, 
    order=orden_barrio
)
plt.title('Frecuencia de registros por tipo de riesgo y barrio')
plt.show()
```

- Hay muchas diferencias entre los barrios, esta variable va a ser importante.
- No hay una tendencia a que los barrios con mayor frecuencia de alto riesgo sean los de menor poder adquisitivo
- Hay algunos Barrios que tienen muy poca frecuencia de alto riesgo como: Pocitos, es justamente uno de los barrios mas densamente poblados
- Unión es otro de los barrios más densamente poblados pero que tiene la mayor frecuencia de alto riesgo de delitos

```python
# Gráfico de Porcentaje de distribucion de riesgo por variable categorica

# Eliminamos la variable Ba

# Create a grid with enough rows and 2 columns for the crosstab analysis
fig, axes = plt.subplots(nrows=(len(cat_features) // 2 + 1), ncols=2, figsize=(16, 40))
axes = axes.flatten()  # Flatten axes for easy indexing

palette_name = "YlGnBu"  # Select a palette

# Loop through the categorical variables
for i, col in enumerate(cat_features):
    # Compute the crosstab (percentage within each category)
    crosstab_data = pd.crosstab(df[col], df['RIESGO'], normalize='index') * 100
    
    # Generate colors based on the number of columns in crosstab_data
    colors = sns.color_palette(palette_name, n_colors=len(crosstab_data.columns))
    
    # Plot the stacked bar chart
    crosstab_data.plot(kind='bar', stacked=True, ax=axes[i], color=colors)
    axes[i].set_title(f"{col} vs. RIESGO", fontsize=18)
    axes[i].set_ylabel("Percentage (%)", fontsize=14)
    axes[i].set_xlabel(col, fontsize=14)
    axes[i].tick_params(axis='x', rotation=45, labelsize=10)
    axes[i].tick_params(axis='y', labelsize=14)

# Remove unused subplots
for j in range(len(cat_features), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
fig.suptitle("Porcentaje de Distribución de Riesgo por Variables Categóricas", fontsize=20, y=1.02)
plt.show()
```

- En general no hay grandes diferencias en la proporción de alto y bajo riesgo en las variables categóricas observando los valores para todos los barrios.
- Hay un pequeño pico de alto riesgo en febrero y marzo y una bajada mínima en junio y julio.
- Si el feriado es No laborable hay una pequeña disminucion del porcentaje de riesgo alto
- Hay unas diferencias mayores en cuanto al feriado, los feriado por asuncion presidencial hay una mayor proporcion de riesgo alto y los feriados de Navidad una menor proporcion de riesgo alto.

##### 2.2.2. VARIABLES CONTÍNUAS

```python
# Observamos correlaciones entre variables numéricas

# Calculamos correlaciones
corr = df[num_features].corr().round(2)

f, ax = plt.subplots(figsize=(12, 4))

# Mascara para ocultar la mitad de la matriz
mask = np.triu(np.ones_like(corr, dtype=bool))
np.fill_diagonal(mask, False)

# Creamos el grafico de correlaciones
sns.heatmap(corr, vmin=-1, vmax=1, annot=True, fmt=".3f", annot_kws={"size": 7}, cmap='BrBG', mask=mask)

plt.title ("Matriz de corelación de Pearson", y=1.1, fontsize=13)
plt.show()
```

- Hay alta coorelacion en las variables que miden temperatura y sensación térmica.
- Se podría mantener solamente TEMPMAX y de esa forma se elimina la multicolinealidad.

```python
# Valores promedio de variables numéricas por tipo de riesgo
df_num_grop = df.groupby(['RIESGO'])[num_features].mean()
df_num_grop.T
```

- La principal diferencia es que en promedio cuando hay bajo riesgo de delitos, el volumen de precipitaciones es un poco mayor que cuando el riesgo es alto.
- Si la poblacion del barrio es menor, hay más frecuencia de alto riesgo
