```python
# Seleccionamos las columnas  que necesitamos
df_sel = df[['datetime', 'tempmax', 'tempmin', 'temp', 'feelslikemax',
        'feelslikemin', 'feelslike', 'humidity', 'precip', 'windgust', 'windspeed', 'winddir', 'sealevelpressure', 'visibility', 'severerisk', 'conditionClear', 'conditionPartiallyCloudy', 'conditionOvercast', 'conditionRain',]].copy() 
```

- Direccion de vientos según ángulo
- https://aemetblog.es/2016/03/29/el-lenguaje-de-los-meteorologos-los-vientos/

```python
# Creamos la columna windirCat con categorias segun direccion del viento Norte, Sur, Este, Oeste

# Función para asignar dirección del viento
def asignarDir (angulo):
    if 0 <= angulo < 45 or 315 <= angulo <= 360:
        return 'Norte'
    elif 45 <= angulo < 135:
        return 'Este'
    elif 135 <= angulo < 225:
        return 'Sur'
    elif 225 <= angulo < 315:
        return 'Oeste'

# Aplicamos la función a la columna winddir
df_sel['windirCat'] = df_sel['winddir'].apply(asignarDir)
```

```python
# Chequeamos columna creada
df_sel.windirCat.describe()
```

```python
# Eliminamos columna windir
df_sel.drop('winddir', axis=1, inplace=True)
```

```python
# Crear nueva columna de fecha solo con dia, mes y año para hacer merge con dataset de delitos agrupados por fecha
df_sel['fecha'] = df_sel['datetime'].dt.date
```

- Los dias sin riesgo (severerisk) tienen datos NaN

```python
# Transformar valors NaN en 0 en columna severerisk
df_sel['severerisk'] = df_sel['severerisk'].fillna(0)
```

```python
# Eliminamos el duplicado en la fecha
df_sel = df_sel.drop_duplicates(subset=['fecha'])
```
